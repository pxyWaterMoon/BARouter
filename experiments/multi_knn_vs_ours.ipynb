```python
from src.algorithms.run_batch_AUPD import main
import os
import yaml
import json

def load_config(config_path):
    # Check if the config file exists
    if not os.path.exists(config_path):
        raise FileNotFoundError(f"Configuration file not found: {config_path}")
    # Load the configuration from the YAML file
    with open(config_path, "r") as file:
        config = yaml.safe_load(file)
    return config
```

```python
config = load_config("src/configs/mf_AUPD_routerbench.yaml")

main(config)
```

```python
B_list = range(400,1501,100)
mu_list = [0.3]
avg_r_list = {}

config = load_config("src/configs/carrot_sprout.yaml")
name = "carrot_0.3"
for val in mu_list:
    config["agent"]["mu"] = val
    title = f"{name}_mu={val}"
    avg_r_list[title]=[]
    for B in B_list:
        print(f"Running system with budget: {B}")
        config["budget"] = B
        config["project_name"] = f"{name}_mu{val}_B{B}"
        r = main(config)
        print(f"Completed run with budget {B}, total reward: {r}")
        print("-" * 50)
        avg_r_list[title].append(r)
print(avg_r_list)

with open(f"./outputs/{name}_batch_results.json", "w") as f:
    json.dump(avg_r_list, f)
```
