```python
import matplotlib.pyplot as plt
import numpy as np
import json
import os

os.chdir("../")
print("Current work path:", os.getcwd())
```

```python
datapath = "outputs/predictors/1006_rb.json"
# datapath = "outputs/exp1/PD.json"
# datapath = "./outputs/predictors/B=1600.json"

data = json.load(open(datapath, 'r'))
```

```python
# alg_list = ["carrot2", "ratio", "cons", "AUPD_exp"]
alg_list = ["ratio"]
pre_list = ["olknn", "kmeans", "xgbregressor"]
budget_list = [2000,5000,10000]
for budget in budget_list:
    for pre in pre_list:
        for alg in alg_list:
            key = f"{alg}_{pre}_B={budget}"
            d = np.array(data[key])
            print(f"{d.mean():.3f} {d.std():.3f}",key)
```
